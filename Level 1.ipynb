<h2 style="color:black;">1) Data Import</h2>

```python
# Import Required Libraries
import pandas as pd
```

```python
# Load the Dataset
df= pd.read_csv('iris.csv')
```

```python
# Preview the Data
df.head()
```

<h2 style="color:black;">2) Checking for missing values</h2>

```python
# Check for Missing Values
df.isnull().sum()
```

<h2 style="color:black;">3) Checking and handling duplicates</h2>

```python
# Check for Duplicates
df.duplicated().sum()
```

```python
# Remove Duplicated rows
df.drop_duplicates(inplace=True)
```

```python
# Checking for Duplicates
df.duplicated().sum()
```

<h2 style="color:black;">4) Uniform Formatting for Species Names</h2>

```python
#Standardize Column Formats
#Convert species names to lowercase and strip whitespace
df['species']=df['species'].str.lower().str.strip()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2 style="color:black;">1) Summary Statistics </h2>

```python
# Summary Statistics
df.describe()
```

<h2 style="color:black;">2) Visualizing distribution of features</h2>

```python
Features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(1, 4, figsize=(16, 4))  # 1 row and 4 columns

for i in range(4):
    sns.histplot(df[features[i]], kde=True, ax=axes[i])
    axes[i].set_title(features[i])
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
plt.show()
```

<h2 style="color:black;">3) Visualizing with box plot </h2>

```python
# Boxplots by Species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(1, 4, figsize=(16, 4))  # 1 row, 4 columns
for i in range(4):
    sns.boxplot(
        x="species", 
        y=features[i], 
        data=df, 
        hue="species",
        ax=axes[i]
    )
    axes[i].set_title(features[i])
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
plt.show()
```

<h2 style="color:black;">4) Visualizing with the scatter plot  </h2>

```python
# Scatter Plot Matrix
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=df)
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

<h2 style="color:black;">5) Finding Correlations between numerical features </h2>

```python
# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[features].corr(), annot=True, cmap='Blues')
plt.title('Feature Correlation Matrix')
plt.show()
```
